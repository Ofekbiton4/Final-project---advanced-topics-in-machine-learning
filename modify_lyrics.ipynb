### Step 2 - Cleaning the columns and preparing the data for further work

```python
df['name'] = df['name'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
df['artist'] = df['artist'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
```

```python
df['year'] = df['year'].astype(str)
df['year'] = df['year'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
df['year'] = df['year'].astype(int)
```

```python
df_deduped = df.drop_duplicates(subset=['name', 'artist'])
```

```python
def get_decade(year):
    if year >= 2000:
        decade_start = year // 10 * 10
        return f"{decade_start}s"
    else:
        decade_start = (year // 10) * 10
        return f"{str(decade_start)[2:]}s"

df['decade'] = df['year'].apply(get_decade)
```

Cleaning function for the lyrics column - removing all signs, tokenization and preparation for further analysis

```python
def preprocess_lyrics(text):
    # Tokenization
    tokens = word_tokenize(text)
    # Lowercasing
    tokens = [token.lower() for token in tokens if token.isalpha()]  # Ensuring token is alphabetic
    # Removing stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    # Stemming
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(token) for token in tokens]
    return tokens

df['pre_lyrics'] = df['lyrics'].apply(preprocess_lyrics)
```

```python
sia = SentimentIntensityAnalyzer()

df['Sentiment_Score'] = df['lyrics'].apply(lambda x: sia.polarity_scores(x)['compound'])

def categorize_sentiment(score):
    if score > 0.05:
        return 'Positive'
    elif score < -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment_Category'] = df['Sentiment_Score'].apply(categorize_sentiment)
```

```python
def classify_song(lyrics, categories):
    for category, keywords in categories.items():
        if any(keyword in lyrics.lower() for keyword in keywords):
            return category
    return 'Uncategorized'

df['category'] = df['lyrics'].apply(lambda x: classify_song(x, categories))

category_counts = df['category'].value_counts()

plt.figure(figsize=(12, 10))
category_counts_sorted = category_counts.sort_values(ascending=True)
category_counts_sorted.plot(kind='barh', color='skyblue')
plt.title('Distribution of Song Categories Based on Lyrics')
plt.xlabel('Number of Songs')
plt.ylabel('Category')
plt.show()
```

### Part 2 : Segmentation of songs and playlists according to toxicity and offensive scores

```python
# Initialize Detoxify model
model = Detoxify('original')

# Offensive Language Detection
def get_toxicity_score(text):
    results = Detoxify('original').predict(text)
    return results['toxicity']

df['toxicity_score'] = df['lyrics'].apply(get_toxicity_score)
df['offensive'] = df['toxicity_score'].apply(lambda x: 'Yes' if x > 0.5 else 'No')
```

```python
artist_agg = df.groupby('artist').agg(average_popularity=('popularity', 'mean'),
                                       average_toxicity_score=('toxicity_score', 'mean')).reset_index()

top_artists = artist_agg.sort_values(by='average_popularity', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.scatter(top_artists['average_popularity'], top_artists['average_toxicity_score'], color='blue')
for i, txt in enumerate(top_artists['artist']):
    plt.annotate(txt, (top_artists['average_popularity'].iloc[i], top_artists['average_toxicity_score'].iloc[i]))
plt.title('Average Toxicity Score vs. Popularity for Top 10 Artists')
plt.xlabel('Average Popularity')
plt.ylabel('Average Toxicity Score')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.groupby('year')['toxicity_score'].mean().plot(kind='line', color='purple')
plt.title('Average Toxify Score Over Years')
plt.xlabel('Year')
plt.ylabel('Average Toxify Score')
plt.grid(True)
plt.show()
```

```python
playlist_offensive_percentage = {}
for playlist_id, group in df.groupby('playlist_id'):
    total_songs = len(group)
    offensive_songs = (group['offensive'] == 'Yes').sum()
    offensive_percentage = offensive_songs / total_songs * 100
    playlist_offensive_percentage[playlist_id] = offensive_percentage

prohibited_playlists = []
for playlist_id, offensive_percentage in playlist_offensive_percentage.items():
    if offensive_percentage > 30:
        prohibited_playlists.append(playlist_id)

print("Prohibited Playlists for Under 18s:")
for playlist_id in prohibited_playlists:
    print("Playlist ID:",playlist_id)
```
